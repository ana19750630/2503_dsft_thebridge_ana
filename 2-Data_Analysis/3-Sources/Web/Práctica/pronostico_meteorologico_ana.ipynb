```python
weather = requests.get(url2)
```

```python
weather
```

```python
weather.content
```

```python
data = weather.json() ##Analizar las respuestas formato Json
```

```python
data
```

```python
hora = data ["hourly"]["time"]
temperatura = data ["hourly"]["temperature_2m"]
```

```python
# Crear un nuevo diccionario con la hora y temperatura
diccionario = {"hora": hora, "temperatura": temperatura}
diccionario
```

```python
##df = pd.DataFrame(hora, temperatura)
```

```python
df = pd.DataFrame({"hora": hora, "temperatura": temperatura})
```

```python
df
```

```python
# Crear el gráfico de líneas
plt.figure(figsize=(10, 6))
plt.plot(df['hora'], df['temperatura'], marker='o', color='b', label="Temperatura")

# Añadir título y etiquetas
plt.title('Pronóstico de Temperaturas a lo Largo del Día', fontsize=14)
plt.xlabel('Hora', fontsize=12)
plt.ylabel('Temperatura (°C)', fontsize=12)

# Mejorar la visualización de las horas en el eje X
plt.xticks(rotation=45)

# Mostrar la leyenda
plt.legend()

# Mostrar la cuadrícula para facilitar la lectura
plt.grid(True)

# Ajustar el layout para no cortar etiquetas
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
