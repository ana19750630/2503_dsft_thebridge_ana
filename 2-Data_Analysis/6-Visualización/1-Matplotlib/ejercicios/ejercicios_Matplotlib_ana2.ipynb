2. Sabemos que podemos pintar gráficas de dos formas: la figura incluye los ejes o teniendo figura y ejes por separado.

Usando solo una figura, usa numpy para los valores del eje X entre 0 y 5. 

Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). 

Elige la precisión en el eje X que desees.  

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar valores para el eje X entre 0 y 5, con 100 puntos (alta precisión)
x = np.linspace(0, 5, 100)

# Calcular las rectas
y1 = 3 * x  # recta con pendiente positiva de 3 pasando por (0,0)
y2 = -3 * x - 5  # recta con pendiente negativa de 3 pasando por (0,-5)

# Crear una figura
plt.figure(figsize=(10, 5))

# Graficar la primera recta (a la izquierda)
plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, primer gráfico
plt.plot(x, y1, label='y = 3x', color='blue')
plt.title('Recta con pendiente positiva')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Graficar la segunda recta (a la derecha)
plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, segundo gráfico
plt.plot(x, y2, label='y = -3x - 5', color='red')
plt.title('Recta con pendiente negativa')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Mostrar los gráficos
plt.tight_layout()  # Ajusta el espaciado para evitar solapamientos
plt.show()
```

5. Usando una figura Y EJES POR SEPARADO, usa numpy para los valores del eje X entre 0 y 5. Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). Elige la precisión en el eje X que desees.  

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar valores para el eje X entre 0 y 5, con 100 puntos (alta precisión)
x = np.linspace(0, 5, 100)

# Calcular las rectas
y1 = 3 * x  # recta con pendiente positiva de 3 pasando por (0,0)
y2 = -3 * x - 5  # recta con pendiente negativa de 3 pasando por (0,-5)

# Crear una figura
fig = plt.figure(figsize=(10, 5))

# Crear los ejes para la primera gráfica (a la izquierda)
ax1 = fig.add_subplot(121)  # 1 fila, 2 columnas, primer gráfico
ax1.plot(x, y1, label='y = 3x', color='blue')
ax1.set_title('Recta con pendiente positiva')
ax1.set_xlabel('eje X')
ax1.set_ylabel('eje Y')
ax1.set_xlim(0, 5)  # Fijar el eje X entre 0 y 5
ax1.set_ylim(-15, 15)  # Fijar el eje Y entre -15 y 15
ax1.grid(True)
ax1.legend()

# Crear los ejes para la segunda gráfica (a la derecha)
ax2 = fig.add_subplot(122)  # 1 fila, 2 columnas, segundo gráfico
ax2.plot(x, y2, label='y = -3x - 5', color='red')
ax2.set_title('Recta con pendiente negativa')
ax2.set_xlabel('eje X')
ax2.set_ylabel('eje Y')
ax2.set_xlim(0, 5)  # Fijar el eje X entre 0 y 5
ax2.set_ylim(-15, 15)  # Fijar el eje Y entre -15 y 15
ax2.grid(True)
ax2.legend()

# Ajustar el layout de la figura para evitar solapamientos
plt.tight_layout()

# Mostrar la figura
plt.show()
```

11. Prueba a invertir los ejes con ax.invert_axis()
Necesitarás tener los ejes disponibles fuera de la figura

```python
import matplotlib.pyplot as plt

# Nombres de los amigos
amigos = ['Ana', 'Luis', 'Pedro', 'Sofía', 'Carmen']

# Alturas de los amigos en centímetros
alturas = [160, 180, 175, 190, 170]

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(5, 3))  # Crear la figura y los ejes

# Crear un gráfico de barras horizontal con las barras en color verde
ax.barh(amigos, alturas, color='green')

# Añadir título y etiquetas
ax.set_title('Altura de un grupo de amigos')
ax.set_xlabel('Altura (cm)')
ax.set_ylabel('Amigos')

# Eliminar la escala del eje Y
ax.set_yticks([])  # Elimina las marcas y etiquetas del eje Y

# Invertir los ejes
ax.invert_yaxis()  # Invierte el eje Y

# Mostrar el gráfico
plt.show()
```
